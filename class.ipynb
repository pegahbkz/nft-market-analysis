```python
import pandas as pd
```

```python
data = pd.read_csv('jan23.csv', low_memory=False)
data
```

```python
data['asset_asset_contract_asset_contract_type_semi'] = data['asset_asset_contract_asset_contract_type_semi'].replace({'TRUE': 1, 'FALSE': 0})
data['asset_collection_is_rarity_enabled'] = data['asset_collection_is_rarity_enabled'].replace({'TRUE': 1, 'FALSE': 0})
data['asset_collection_is_creator_fees_enforced'] = data['asset_collection_is_creator_fees_enforced'].replace({'TRUE': 1, 'FALSE': 0})
```

```python
data['asset_description'] = data['asset_description'].astype(str)
data['asset_asset_contract_description'] = data['asset_asset_contract_description'].astype(str)
```

```python
new_column = []

for index, row in data.iterrows():
    if row['asset_description'] == '1' or row['asset_asset_contract_description'] == '1':
        new_column.append(1)
    else:
        new_column.append(0)
        #data3.append(row)

data['generator'] = new_column
```

```python
data = data.drop(['asset_description','asset_asset_contract_description'], axis=1)
```

```python
data = data.join(pd.get_dummies(data.asset_asset_contract_chain_identifier)).drop(['asset_asset_contract_chain_identifier'], axis=1)
data = data.join(pd.get_dummies(data.asset_collection_display_data_card_display_style)).drop(['asset_collection_display_data_card_display_style'], axis=1)
data = data.join(pd.get_dummies(data.asset_collection_safelist_request_status)).drop(['asset_collection_safelist_request_status'], axis=1)
data = data.join(pd.get_dummies(data.payment_token_symbol)).drop(['payment_token_symbol'], axis=1)
```

```python
data.info()
```

```python
df2 = pd.DataFrame(columns=data.columns)

# Store the dtypes of original_df columns
dtypes_dict = data.dtypes.to_dict()

# Iterate through the original DataFrame
indices_to_remove = []  # To store indices of rows to be removed
for index, row in data.iterrows():
    if row['generator'] == 1:
        # Add rows where 'a' is equal to 1 to df2
        df2 = df2.append(row, ignore_index=True)
        indices_to_remove.append(index)  # Store indices to remove

# Explicitly cast the types in df2 to match original_df
for col, dtype in dtypes_dict.items():
    df2[col] = df2[col].astype(dtype)
    
data = data.drop(indices_to_remove)
```
